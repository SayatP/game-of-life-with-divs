```python
arr = [[j for j in range(10*i+1,10*i+11)] for i in range(10)]
```

```python
offset = (0, 1, -1)
indices = [(i, j) for i in offset for j in offset]
indices.remove((0,0))

def find_neighbours(matrix, idx):
    neighbors = []

    r = (idx-1)//10
    c = idx%10-1
    for rec in indices:
        try:
            row = r + rec[0]
            col = c + rec[1]
            if 0 <= row and 0 <= col: neighbors.append(matrix[row][col])
        except IndexError:
            continue
    return neighbors
```
